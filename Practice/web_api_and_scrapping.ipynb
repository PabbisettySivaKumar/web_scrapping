### #Omicron Tweets

```python
query = "#omicron"


data = {
    'text':[],
    'created_at':[],
    'author_id':[]
}

for tweet in tweepy.Paginator(client.search_recent_tweets, query=query,
                             tweet_fields=['created_at','author_id']).flatten(limit=100):
    data['text'].append(tweet.text)
    data['created_at'].append(tweet.created_at)
    data['author_id'].append(tweet.author_id)
    
```

### Parse the HTML

```python
def extract_categories(baseurl):
    categories = {}
    r = requests.get(baseurl)
    soup =bs(r.content)
    cat_list = soup.find("ul", class_="nav-list").ul.find_all("li")
    
    for li in cat_list:
        categories.update({li.text.strip(): baseurl + li.a['href']})
    return categories
        
```

```python
extract_categories(baseurl)
```
